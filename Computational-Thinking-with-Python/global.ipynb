```python
# importando as bibliotecas necessárias
import pandas as pd
from datetime import datetime
```

```python
# criando um dicionário para salvar os cadastros 
cadastros = {
    "Nome do paciente":[],
    "Data de nascimento": [],
    "CPF do paciente":[],
    "Email do paciente":[],
    "Senha do paciente":[],
    "Telefone do paciente":[]
}
```

```python
# criando um segundo dicionário para salvar os prontuários dos pacientes
prontuarios = {
    "Paciente":[],
    "Email para contato": [],
    "Data de nascimento": [],
    "Data de acesso":[],
    "Sintomas":[],
    "Medicamentos":[],
    "Alergias":[],
    "Doenças":[],
}
```

```python
# criando função para cadastrar pacientes
def cadastrar_paciente():
# criando inputs com nome, data de nascimento, cpf, email, senha e telefone
    nome = input("Nome")
    data = input("Data de nascimento")
    cpf = input("CPF")
    email = input("E-mail")
    senha = input("Senha")
    telefone = input("Telefone")
# criando estrutura if else que verifica se o cpf cadastrado já está contido no dicionário de cadastros
    if cpf not in cadastros["CPF do paciente"]:
# caso o cpf inserido não esteja contido no dicionário, adicione todos os inputs do usuário no dicionário        
        cadastros["Nome do paciente"].append(nome)
        cadastros["Data de nascimento"].append(data)
        cadastros["CPF do paciente"].append(cpf)
        cadastros["Email do paciente"].append(email)
        cadastros["Senha do paciente"].append(senha)
        cadastros["Telefone do paciente"].append(telefone)
# caso contrário, imprima uma mensagem indicando que o paciente já está cadastrado        
    else:
        print("\nEsse CPF já está cadastrado \nFaça o login para continuar")

```

```python
# criando função para acessar todos os cadastros
def acessar_cadastros():
    # criando variável df_cadastros que mostra as informações contidas no dicionário cadastro utilizandoa biblioteca importada
    df_cadastros = pd.DataFrame(cadastros)
    # caso o dicionário esteja vazio, imprima a mensagem "Nenhum cadastro encontrado."
    if len(df_cadastros) == 0:
        print("Nenhum cadastro encontrado.")
    # caso contrário, imprima a variável criada    
    else:
        return df_cadastros
    
```

```python
# criando listas vazias para salvar as informações do paciente logado
paciente_logado = []
email_logado = []
data_logada = []
# criando uma função para que o paciente efetue login usando email e senha
def login_paciente():
    login = input("Digite o seu email")
    senha = input("Digite sua senha")
# caso o email e senha inseridos pelo paciente estejam cadastrados:
    if login in cadastros["Email do paciente"] and senha in cadastros["Senha do paciente"]:
# imprimindo uma mensagem que indica que o login foi um sucesso
        print("Login efetuado com sucesso!")
# salvando o indice do email do paciente em uma variável index         
        index = cadastros["Email do paciente"].index(login)
# salvando o nome que partilha o mesmo indice no dicionário cadastros na lista vazia paciente        
        paciente_logado.append(cadastros["Nome do paciente"][index])
        email_logado.append(cadastros["Email do paciente"][index])
        data_logada.append(cadastros["Data de nascimento"][index])
# imprimindo uma mensagem de boas vindas para o paciente        
        print(f"Bem vindo {paciente_logado[0]}!")
# caso contrário:
    else:
# imprimindo a mensagem "Usuário ou Senha inválido."         
        print("Usuário ou Senha inválido.")
```

```python
# criando uma função para fazer o logout do paciente
def logout_paciente():
# imprimindo uma mensagem de despedida ao paciente
    print(f"Até mais {paciente_logado[0]}!")
# limpando as listas utilizadas para armazenar os dados do paciente logado
    paciente_logado.clear()
    email_logado.clear()
    data_logada.clear()
```

```python
# criando uma função para registrar prontuários dos pacientes logados
def prontuario():              
# usando biblioteca importada para salvar a data e hora em que foi preenchido o prontuário
    dia_e_horas = datetime.now()
    leitura_dia_e_horas = dia_e_horas.strftime("%d/%m/%Y %H:%M:%S")
# criando campos para que o paciente especifique os detalhes do prontuário
    sintomas = input("Quais sintomas você notou? Quando começaram os sintomas? Descreva de maneira detalhada")
    medicamentos = input("Faz uso de medicamentos? Se sim, quais?")
    alergias = input("Possui alergias? Se sim, quais?")
    doencas = input("Está fazendo tratamento de alguma doença? Se sim, quais?")
# adicionando os dados do prontuário ao dicionário prontuários 
    prontuarios["Paciente"].append(paciente_logado[0])
    prontuarios["Email para contato"].append(email_logado[0])
    prontuarios["Data de nascimento"].append(data_logada[0])  
    prontuarios["Data de acesso"].append(leitura_dia_e_horas)
    prontuarios["Sintomas"].append(sintomas)
    prontuarios["Medicamentos"].append(medicamentos)
    prontuarios["Alergias"].append(alergias)
    prontuarios["Doenças"].append(doencas)
 

```

```python
# criando função para acessar todos os prontuários
def acessar_prontuarios():
# criando variável df_prontuarios que mostra as informações contidas no dicionário prontuarios utilizando a biblioteca importada
    df_prontuarios = pd.DataFrame(prontuarios)
# caso o dicionário esteja vazio, imprima a mensagem "Nenhum prontuário encontrado."
    if len(df_prontuarios) == 0:
        print("Nenhum prontuário encontrado.")
# caso contrário, imprima a variável criada        
    else:
        return df_prontuarios
```

```python
# criando função para verificar todos os prontuários de um paciente
def prontuarios_do_paciente():
# atualizando a variável df_prontuarios 
    df_prontuarios = pd.DataFrame(prontuarios)
# perguntando qual paciente deseja se verificar    
    paciente_desejado = input("Qual paciente você deseja verificar?")
# caso o paciente desejado tenha um prontuario:
    if paciente_desejado in prontuarios["Paciente"]:
# criando uma lista vazia para armazenar os indices de todas as vezes em que o paciente desejado aparece no dicionario prontuario        
        indices = []
# criando um loop que percorre a lista "Paciente" no dicionario com dois argumentos, index e item       
        for index, item in enumerate(prontuarios["Paciente"]):
# caso o item seja igual ao paciente desejado, adicione o indice correspondente na lista indices            
            if item == paciente_desejado:
                indices.append(index)
# criando um loop que percorre todos os indices salvos na lista indices                
        for index in indices:
# imprimindo os items do dicionario prontuarios que partilham dos indices salvos na lista indices            
            display(df_prontuarios.iloc[index])
# caso contrário            
    else:
# imprimindo mensagem "Não foi encontrado o paciente solicitado" 
        print("Não foi encontrado o paciente solicitado")

   
```
