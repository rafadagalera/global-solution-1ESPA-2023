```python
import sympy 
from sympy.plotting import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
#GRAFICO 1 - Relacionando o tempo de espera e a satisfação do paciente de maneira hipotetica.
# OBS - O objetivo é demonstrar que são grandezas inversamente proporcionais, ou seja , que conforme uma aumenta a outra diminui independente de seus valores aplicaveis na vida real.

def relST(z):
    return 1/z

# Gerando dados para o eixo x (de 1 a 10)
x_values = range(1, 11)

# Calculando os valores correspondentes para o eixo y usando a função relST
y_values = [relST(value) for value in x_values]

# Plotando o grafico
plt.plot(x_values, y_values)
plt.ylabel('Tempo de espera em horas')
plt.xlabel('Satisfação do paciente')
plt.title('Relação entre Tempo de espera e Satisfação do paciente')
plt.show()
```

```python


def relCH(complexidade):
    # Ajustando a relação para refletir que maior complexidade significa maior distância
    return complexidade ** 2

# Gerando dados para o eixo x (de 2 a 4)
x_values = [val / 10 for val in range(20, 41)]  # Mapeando de 20 a 40 para 2.0 a 4.0

# Calculando os valores correspondentes para o eixo y usando a função relCH
y_values = [relCH(value) for value in x_values]

# Plotando o gráfico
plt.plot(x_values, y_values)
plt.ylabel('Distancia a ser percorrida KM')
plt.xlabel('Complexidade do atendimento')
plt.title('Relação entre Complexidade do Atendimento e Distância Percorrida')
plt.show()

```
